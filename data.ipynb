# Amazon Book Data Analysis
In this python data analysis, we are taking an Amazon Book list and comparing their price to their rating to see if we can find a trend, or no trend. The goal is to see whether or not the price of the book implies that the quality of the book's content is superior to one of lesser cost.

We will be using python to grab the data and use it to make a scatter plot

### Importing the necessary packages

```python
import csv
import matplotlib.pyplot as plt 
import matplotlib as plot
```

### Reading in the data from the .csv file

```python
with open("bestsellers with categories.csv", "r", encoding="utf-8") as file:
    book_info = csv.DictReader(file)
    data_list = []
    for r in book_info:
        data_list.append(r)
```

### Reference Object


```python
data_list[0]
```

### The x and y axis
Naming two variables that will hold the information of our scatter plot and filling those arrays with information from our .csv

```python
x_axis_price = []
y_axis_reviews = []

for each in data_list:
    x_axis_price.append(int(each["Price"]))
    y_axis_reviews.append(float(each["User Rating"]))
```

### Plotting the Results in a Scatter Plot

```python
plt.scatter(x_axis_price, y_axis_reviews)
plt.xlabel('Price')
plt.ylabel('Average Review Score')
plt.title('Amazon Book price compared to its rating')
plt.show()
```

# Results

We can see from this data chart that when the books rating is compared to price, most reviews are concentrated within the $0 - $30 dollar range, we have outliers outside of this that may skey the data. Let's take a closer look to see if there is a trend within the range of $0 - $35 dollars

```python
x_axis_price = []
y_axis_reviews = []

for each in data_list:
    x_axis_price.append(int(each["Price"]))
    y_axis_reviews.append(float(each["User Rating"]))
```

```python
print(len(x_axis_price))
print(len(y_axis_reviews))
```

### Getting rid of the outliers

```python
smaller_x = []
smaller_y = y_axis_reviews

for num in x_axis_price:
    if num <= 35:
        smaller_x.append(num)

n = 24

del smaller_y[-n:]

print(len(smaller_x))
print(len(smaller_y))
```

```python
plt.scatter(smaller_x, smaller_y)
plt.xlabel('Price')
plt.ylabel('Average Review Score')
plt.title('Amazon Book price compared to its rating')
plt.show()
```
